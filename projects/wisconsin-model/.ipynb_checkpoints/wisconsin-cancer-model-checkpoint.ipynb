Available at https://archive.ics.uci.edu/ml/datasets/Breast+Cancer+Wisconsin+(Diagnostic)
 
The data includes diagnosis of the tumor with  M  meaning cancer (malignant) and  B  no cancer (benign), and 10 features, describing physical properties of cell nuclei from biopsy samples. Each feature is represented three times, once for mean, once for standard error, and once for the worst values. We will predict diagnosis based on this data.

### Explore the data

Loading the data. Drop id or ignore it in the rest of the analysis.

```python
import pandas as pd
import numpy as np
from scipy.stats.stats import pearsonr
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR
from sklearn.svm import SVC


wbcd = pd.read_csv("wdbc.csv")
wbcd = wbcd.drop(columns="id")

# 0: benign, 1: malignant
wbcd["diagnosis"] = [1 if i == "M" else 0 for i in wbcd["diagnosis"]]
```

Summary table showing means, ranges, and number of missings for each variable and (Pearson) correlation between the diagnosis and corresponding feature.

```python
summary = pd.DataFrame(columns=['name','mean', 'range (max-min)', 'nmissing', 'correlation'])
for c in wbcd.drop(columns="diagnosis").columns:
    m = np.mean(wbcd[c])
    r = max(wbcd[c]) - min(wbcd[c])
    nm = wbcd[c].isna().sum()
    cor = pearsonr(wbcd[c], wbcd["diagnosis"])
    summary = summary.append({'name':c,'mean':m,'range (max-min)':r,'nmissing':nm,'correlation':cor},ignore_index=True)
    
summary.head()
```

A number of scatterplots exploring the relationship between features and the diagnosis.

```python
ax1=plt.subplot(3, 2, 1)
plt.scatter(wbcd["diagnosis"], wbcd["concpoints.worst"])
ax1.set(xlabel='diagnosis', ylabel='concpoints.worst')

ax2=plt.subplot(3, 2, 2)
plt.scatter(wbcd["diagnosis"], wbcd["perimeter.worst"])
ax2.set(xlabel='diagnosis', ylabel='perimeter.worst')

ax3=plt.subplot(3, 2, 3)
plt.scatter(wbcd["diagnosis"], wbcd["radius.worst"])
ax3.set(xlabel='diagnosis', ylabel='radius.worst')

ax4=plt.subplot(3, 2, 4)
plt.scatter(wbcd["diagnosis"], wbcd["fracdim.se"])
ax4.set(xlabel='diagnosis', ylabel='fracdim.se')

ax5=plt.subplot(3, 2, 5)
plt.scatter(wbcd["diagnosis"], wbcd["texture.se"])
ax5.set(xlabel='diagnosis', ylabel='texture.se')

ax6=plt.subplot(3, 2, 6)
plt.scatter(wbcd["diagnosis"], wbcd["fracdim.se"])
ax6.set(xlabel='diagnosis', ylabel='fracdim.se')

plt.tight_layout()
plt.show()
```

### Decision Boundary

#### kNN Case

Prediction of the diagnosis on a grid that covers the range of explanatory variables. Using kNN with k = 3..7. 


```python
X1 = np.random.normal(size=100)
X2 = np.random.normal(size=100)
Y = X1 + X2 + np.random.normal(size=100) > 0
X = np.stack((X1, X2), axis=1)
# m = LogisticRegression().fit(X, Y)
## Now greate the regular grid
ex1 = np.linspace(X1.min(), X1.max(), 100)
ex2 = np.linspace(X2.min(), X2.max(), 100)
xx1, xx2 = np.meshgrid(ex1, ex2)
g = np.stack((xx1.ravel(), xx2.ravel()), axis=1)
len(g)

```

```python
scaler = StandardScaler()
scaler.fit(wbcd[["concpoints.worst", "perimeter.worst"]])

X = scaler.transform(wbcd[["concpoints.worst", "perimeter.worst"]])
y = wbcd["diagnosis"]

m = KNeighborsClassifier(n_neighbors=5)
m.fit(X, y)

## Now greate the regular grid
ex1 = np.linspace(X[:,0].min(), X[:,0].max(), 100)
ex2 = np.linspace(X[:,1].min(), X[:,1].max(), 100)
xx1, xx2 = np.meshgrid(ex1, ex2)

g = np.stack((xx1.ravel(), xx2.ravel()), axis=1)

hatY = m.predict(g).reshape(100,100)
```

Actual data and the decision boundary on the same plot. 

```python
plt.imshow(hatY, extent=(X[:,0].min(), X[:,0].max(), X[:,1].min(), X[:,1].max()), interpolation='none', origin='lower', alpha=0.3)
plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```

kNN appears to generate a decision boundary that matches the data. Noise is picked up to a relatively high degree, but not completely. Because the prediction data does not span the entire range of both variables, some speculation about the accuracy of the decision boundary needs to be made.

#### Feature Engineering

```python
scaler = StandardScaler()

wbcd["concavity.worst.3"] = [i**3 for i in wbcd["concavity.worst"]]
wbcd["perimeter.worst.5"] = [i/2 for i in wbcd["perimeter.worst"]]

scaler.fit(wbcd[["concavity.worst.3", "perimeter.worst.5"]])
X = scaler.transform(wbcd[["concavity.worst.3", "perimeter.worst.5"]])
y = wbcd["diagnosis"]
```

```python
m = LogisticRegression().fit(X, y)
```

Decision boundary plot.

```python
## Now greate the regular grid
ex1 = np.linspace(X[:,0].min(), X[:,0].max(), 100)
ex2 = np.linspace(X[:,1].min(), X[:,1].max(), 100)
xx1, xx2 = np.meshgrid(ex1, ex2)

g = np.stack((xx1.ravel(), xx2.ravel()), axis=1)

hatY = m.predict(g).reshape(100,100)
plt.figure(figsize=(20,20))
plt.imshow(hatY, extent=(X[:,0].min(), X[:,0].max(), X[:,1].min(), X[:,1].max()), interpolation='none', origin='lower', alpha=0.3)
plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```

Overall, this model still appears to be more-or-less accurate. The boundry does not seem to fit as well, and overfitting appears to be an issue with the addition of heavily engineered variables. Additionally, the inconsistent engineering resulted in a skewed matrix, which is expected. Perhaps less adjustment of variables would result in less skew nad overfitting.

### ROC curve: which estimator is the best

Comparing estimators in two ways: first using accuracy, precision, recall; and thereafter with ROC curve.

Split data into training-validation parts.

```python
from sklearn.model_selection import train_test_split

scaler = StandardScaler()

wbcd["concavity.worst.3"] = [i**1.3 for i in wbcd["concavity.worst"]]
wbcd["perimeter.worst.5"] = [i**1.2 for i in wbcd["perimeter.worst"]]

scaler.fit(wbcd[["concavity.worst.3", "perimeter.worst.5"]])
X = scaler.transform(wbcd[["concavity.worst.3", "perimeter.worst.5"]])
y = wbcd["diagnosis"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1)
```

Good models from the previous sections.

```python
# There are not so many kernels so no need to loop through
# linear, poly, rbf, sigmoid, precomputed
m1 = KNeighborsClassifier(n_neighbors=5)
m2 = LogisticRegression()
m3 = SVC(kernel='rbf',C=2.0,probability=True)
```

Each model is trained on the training data. Confusion matrix and compute accuracy, precision, recall, and F-score are presented.

```python
# I assumed that we can import this, and that 'precision',
# 'accuracy', 'recall', and 'F-score' were meant by "algorithms".
from sklearn.metrics import confusion_matrix

m1.fit(X_train,y_train)
m2.fit(X_train,y_train)
m3.fit(X_train,y_train)

x = confusion_matrix(y_test, m1.predict(X_test))
print("kNN:\n",x,"\n")
print("accuracy:",((x[1,1]+x[0,0])/(x[0,0]+x[0,1]+x[1,0]+x[1,1]))) # TP + TN / TP + TN + FP + FN
p = x[1,1]/(x[1,1]+x[0,1])
print("precision:", p) # TP / TP + FP
r = x[1,1]/(x[1,1]+x[1,0])
print("recall:", r) # TP / TP + FN
print("F-score:", 2*(p*r)/(p+r), "\n\n") # 2( p * r / p + r )
x = confusion_matrix(y_test, m2.predict(X_test))
print("Logistic Regression:\n",x,"\n\n")
print("accuracy:",((x[1,1]+x[0,0])/(x[0,0]+x[0,1]+x[1,0]+x[1,1]))) # TP + TN / TP + TN + FP + FN
p = x[1,1]/(x[1,1]+x[0,1])
print("precision:", p) # TP / TP + FP
r = x[1,1]/(x[1,1]+x[1,0])
print("recall:", r) # TP / TP + FN
print("F-score:", 2*(p*r)/(p+r), "\n\n") # 2( p * r / p + r )
x = confusion_matrix(y_test, m3.predict(X_test))
print("SVM:\n",x,"\n")
print("accuracy:",((x[1,1]+x[0,0])/(x[0,0]+x[0,1]+x[1,0]+x[1,1]))) # TP + TN / TP + TN + FP + FN
p = x[1,1]/(x[1,1]+x[0,1])
print("precision:", p) # TP / TP + FP
r = x[1,1]/(x[1,1]+x[1,0])
print("recall:", r) # TP / TP + FN
print("F-score:", 2*(p*r)/(p+r), "\n") # 2( p * r / p + r )
```

Probability that the outcome belongs to a given class (say, "Malignant").

```python
models = [m1.fit(X_train,y_train).predict_proba(X_test),
          m2.fit(X_train,y_train).predict_proba(X_test),
          m3.fit(X_train,y_train).predict_proba(X_test)]
```

Predictions are to be "Malignant" if it's probability is at least as big as the threshold for thresholds between 0 and 1.

```python
model_roc_data = []
for m in models: 
    roc_data = []
    for t in [i/1000 for i in range(1001)]:
        preds = []
        for i in m:
            # Treat predictions to be "Malignant" if it's
            # probability is at least as big as threshold
            if i[1] >= t:
                preds.append(1)
            else:
                preds.append(0)
        c_m = confusion_matrix(y_test, preds)
        
        tpr = (c_m[1,1]/(c_m[1,1]+c_m[1,0]))
        fpr = (c_m[0,1]/(c_m[0,1]+c_m[0,0]))
        
        roc_data.append([fpr, tpr])
        
    model_roc_data.append(roc_data)
```

False positive rate and true positive rate for each threshold value. 

```python
for c in model_roc_data:
    plt.plot([row[0] for row in c],[row[1] for row in c])

plt.text(0.8, 0.8, 'green: kNN')
plt.text(0.8, 0.7, 'blue: LR')   
plt.text(0.8, 0.6, 'yellow: SVC')   
plt.xlabel('FP Rate')
plt.ylabel('TP Rate')
plt.title('ROC Curve')
plt.show()
```

This plot shows the accuracy of the models as the thresholds at which these models perform classification changes. The threshold of classification that results in the best model performance is quite low. This could indicate that small variations in the variables are enough to classify a tumor as malignant. All of these models performed relatively equally when comparing thresholds and FP/TP rates, but SVC was the most accurate at all thresholds. This plot confirms the findings of previous steps. SVC/SVR was consistently more accurate than kNN and Logistic Regression.
